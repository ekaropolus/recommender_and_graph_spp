```python
from neo4j import GraphDatabase
import logging
from neo4j.exceptions import ServiceUnavailable

class App:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        # Don't forget to close the driver connection when you are finished with it
        self.driver.close()

    def find_node(self,id):
        with self.driver.session(database="neo4j") as session:
            result = session.read_transaction(self._find_and_return_node,id)
            return result

    @staticmethod
    def _find_and_return_node(tx,id):
        query = (
            "MATCH (n) "
            "RETURN n as id, $id as group"
        )
        result = tx.run(query,id=id)
        return [(row["id"],row["group"]) for row in result]

    def find_relations(self, source,target, relation):
        with self.driver.session(database="neo4j") as session:
            result = session.read_transaction(self._find_and_return_relations,source,target,relation)
            return result

    @staticmethod
    def _find_and_return_relations(tx,source,target,relation):
        query = (
            "MATCH p = (n)-[r]->(m) "
            #"WHERE n.Country = 'UN' "
            "RETURN p as source, $target as target, 5 as value"
        )
        result = tx.run(query,source=source,target=target,relation=relation)
        return [(row["source"],row["target"],row["value"]) for row in result]

```
